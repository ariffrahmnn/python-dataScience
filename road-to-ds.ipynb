```python
nama_kotor = ["  Arif", "ikhsan  ", "  aMinah  "]
hasil_akhir_nama = []
for nama in nama_kotor:
  nama = nama.strip()
  nama = nama.lower()
  nama = nama.capitalize()
  hasil_akhir_nama.append(nama)
print(hasil_akhir_nama)

```

```python
nama_kotor2 = ["  Arif", "ikhsan  ", "  aMinah  "]
nama_bersih = [nama.strip().capitalize() for nama in nama_kotor2]
print(nama_bersih)
```

```python
data_angka = [12, 55, 60, 45, 80, 23, 100]
data_filter = [data for data in data_angka if data > 50 and data % 2 == 0]
print(data_filter)
```
